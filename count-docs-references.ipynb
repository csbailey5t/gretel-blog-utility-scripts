```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_json("blogposts.json")
df.head()
```

```python
posts_with_docs = df[df["post-body"].str.contains("docs.gretel.ai")]
posts_with_docs.head()
```

```python
doc_links = []
for post in posts_with_docs["post-body"]:
    soup = BeautifulSoup(post)
    for link in soup.find_all("a", href=True):
        if "docs.gretel.ai" in link["href"]:
            doc_links.append(link["href"])
```

```python
doc_links
```

```python
# write set of links to docs to file
with open("doc-links.txt", "w") as f:
    f.write("\n".join(set(doc_links)))
```
